```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# data 읽기
iris=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',header=None)
iris.columns= ['sl','sw','pl','pw','name']
```

```python
#step function for single value
def step_func (threshold, x):
   if x<threshold:
       return 0
   return 1

n_trials = 100
n_test = 10
alpha = 0.01 # learning rate
```

```python
#for sl, sw

X1 = iris.loc[10:89, 'sl'] #필요한 컬럼만 남기고 그중에서도 종류별 각 10개씩 test data는 제외한다
X2 = iris.loc[10:89, 'sw']
X = pd.concat([X1, X2], axis=1) #두 컬럼을 합친다
X = X.to_numpy() #numpy array로 변환

T = iris.loc[10:89, 'name'] #target array 추출
T = T.to_numpy()
T = np.where(T=='Iris-setosa', 1, 0) #setosa는 1, 나머지(versicolor)는 0으로 변환

# set test set
Y11 = iris.loc[0:9, 'sl']
Y12 = iris.loc[90:99, 'sl']
Y21 = iris.loc[0:9, 'sw']
Y22 = iris.loc[90:99, 'sw']
Y1 = pd.concat([Y11,Y12], axis=0)
Y2 = pd.concat([Y21,Y22], axis=0)
Y = pd.concat([Y1,Y2], axis=1)
Y = Y.to_numpy()

T1 = iris.loc[0:9, 'name']
T2 = iris.loc[90:99, 'name']
Test = pd.concat([T1,T2], axis=0)
Test = Test.to_numpy()
Test = np.where(Test=='Iris-setosa', 1, 0)

W=np.random.randn(3) #setting random weight value
hit_array=[]

for i in range(n_trials):
    for j in range(len(X)):
        s=np.dot(X[j],W[1:])+W[0]
        delta=alpha*(T[j]-step_func(0,s))
        W[1:]+=delta*X[j]
        W[0]+=delta
        
    hit =0
    for k in range(len(Y)):
        s=np.dot(Y[k],W[1:])+W[0]
        delta=(Test[k]-step_func(0, s))
        if delta==0:
            hit+=1
    
    hit_array.append(int(hit)/20)

finalhitrate_1 = hit_array[-1]

plt.plot(hit_array)
plt.title('sl,sw')
plt.show()
```
