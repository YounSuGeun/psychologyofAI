```python
#Q3
import numpy as np

a=np.random.randint(low=0, high=9, size=1000000000) #random 한자리수 생성
b= 140070 #학번 지정
count = 0

for i in range(999999994):
    c = str(a[i])+str(a[i+1])+str(a[i+2])+str(a[i+3])+str(a[i+4])+str(a[i+5]) #a의 i번째 숫자로부터 6개의 숫자를 순차적으로 결합
    if b == int(c): # c를 다시 숫자로 변환
        count = count + 1 # 숫자 sequence와 학번이 일치하는 횟수 카운트

print(count)
```

```python
#Q4
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('housing.csv') #read csv file


plt.scatter(df['median_house_value'],df['median_income']) #draw scatter plot
plt.show()

from mpl_toolkits import mplot3d
ax=plt.axes(projection='3d')

x=df['longitude']
y=df['latitude']
z=df['median_house_value'] # set x,y,z as each column

ax.plot3D(x,y,z) # draw 3D plot
plt.show()
```

```python
#4-a
from scipy.stats import pearsonr

first_half = df.sample(frac=0.5) # randomly divide df to half
second_half = df.drop(first_half.index) # remove first_half from original df

#calculate pearson correlation coefficient 
corr1, _ = pearsonr(df['median_house_value'][first_half.index], df['median_income'][first_half.index])
corr2, _ = pearsonr(df['median_house_value'][second_half.index], df['median_income'][second_half.index])

print(corr1, corr2)

#two correlation coefficient are close enough since they are randomly divided and each dataset have similar relationship
```

```python
#4-b

df['label'] = df['total_bedrooms']//df['households'] #set label as number of bedrooms per household


df.groupby('label') #categorize into groups by label


#get sum of each group
c=0
for i in df.groupby('label').size().index:
    sum = df.groupby('label').size()[i]*i
    c = c + sum
    
mean = c/len(df) #divide sum to get mean
print('mean=',mean)

#plot frequency of each label
plt.scatter(df.groupby('label').size().index, df.groupby('label').size())

```

```python
#Q6
class Friend:
    def __init__(self,name,age,sex):
        self.name = name
        self.age = age
        self.sex = sex
    
    #define methods
    def chat(self):
        print("What's up bro. I'm", self.name)
        
    def play(self):
        print("You got any plan for tonight?, I got my friends, they are all", self.age)
    
    def eat(self):
        if self.sex == 'M':
            X='boys'
        elif self.sex=='F':
            X='girls'
        
        print("Let's go for %s' night out" % X)

        

 
F = Friend(name = 'Kevin', age = '25', sex = 'M') # define attributes 

F.chat()
F.play()
F.eat()
```
