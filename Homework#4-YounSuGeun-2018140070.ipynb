```python
#Q1
import numpy as np
import matplotlib.pyplot as plt
    
n_arms=10
n_steps= 2000
n_trials=1000

epsilon=0

Result_array=np.zeros(n_steps)   # Result array for plotting
Result_array2=np.zeros(n_steps)   # Result array for plotting optimal action selection
P_arm=np.random.randn(n_arms)    # Average probability of reward amount

for i in range(n_trials):
    
    Q_arm=np.zeros(n_arms) # reward value for each arm
    N_visits = np.zeros(n_arms) #  arm  
    
    for j in range(n_steps):
        R_arm=np.random.normal(P_arm, 1, n_arms) # generate actual reward average probabilty +/- 1 
        
        if np.random.rand(1) < epsilon: 
            action = np.random.choice(n_arms, 1) #  exploration
        else:
            action = np.argmax(Q_arm) #   Reward   arm  Greedy action
            
            
        if N_visits[action] == 0:
            Q_arm[action] += R_arm[action] #  Q value
        else:
            Q_arm[action] += (R_arm[action] - Q_arm[action])/N_visits[action] #Q value update
            
            
        N_visits[action] += 1
        Result_array[j] += R_arm[action] #  trial Reward 
        
        Result_array2[j] += int(np.argmax(R_arm) == action) # count optimal action selection. Boolean to integer
   

plt.plot(Result_array2/n_trials, color='black')        

#-----------------------------------------------------------------
epsilon=0.001

Result_array=np.zeros(n_steps)   # Result array for plotting
Result_array2=np.zeros(n_steps)   # Result array for plotting optimal action selection
P_arm=np.random.randn(n_arms)    # Average probability of reward amount

for i in range(n_trials):
    
    Q_arm=np.zeros(n_arms) # reward value for each arm
    N_visits = np.zeros(n_arms) #  arm  
    
    for j in range(n_steps):
        R_arm=np.random.normal(P_arm, 1, n_arms) # generate actual reward average probabilty +/- 1 
        
        if np.random.rand(1) < epsilon: 
            action = np.random.choice(n_arms, 1) #  exploration
        else:
            action = np.argmax(Q_arm) #   Reward   arm  Greedy action
            
            
        if N_visits[action] == 0:
            Q_arm[action] += R_arm[action] #  Q value
        else:
            Q_arm[action] += (R_arm[action] - Q_arm[action])/N_visits[action] #Q value update
            
            
        N_visits[action] += 1
        Result_array[j] += R_arm[action] #  trial Reward 
        
        Result_array2[j] += int(np.argmax(R_arm) == action) # count optimal action selection
    
plt.plot(Result_array2/n_trials, color='red')

#-----------------------------------------------------------------------
epsilon=0.1

Result_array=np.zeros(n_steps)   # Result array for plotting
Result_array2=np.zeros(n_steps)   # Result array for plotting optimal action selection
P_arm=np.random.randn(n_arms)    # Average probability of reward amount

for i in range(n_trials):
    
    Q_arm=np.zeros(n_arms) # reward value for each arm
    N_visits = np.zeros(n_arms) #  arm  
    
    for j in range(n_steps):
        R_arm=np.random.normal(P_arm, 1, n_arms) # generate actual reward average probabilty +/- 1 
        
        if np.random.rand(1) < epsilon: 
            action = np.random.choice(n_arms, 1) #  exploration
        else:
            action = np.argmax(Q_arm) #   Reward   arm  Greedy action
            
            
        if N_visits[action] == 0:
            Q_arm[action] += R_arm[action] #  Q value
        else:
            Q_arm[action] += (R_arm[action] - Q_arm[action])/N_visits[action] #Q value update
            
            
        N_visits[action] += 1
        Result_array[j] += R_arm[action] #  trial Reward 
        
        Result_array2[j] += int(np.argmax(R_arm) == action) # count optimal action selection
    
plt.plot(Result_array2/n_trials, color='green')

#---------------------------------------------------
epsilon=0.5

Result_array=np.zeros(n_steps)   # Result array for plotting
Result_array2=np.zeros(n_steps)   # Result array for plotting optimal action selection
P_arm=np.random.randn(n_arms)    # Average probability of reward amount

for i in range(n_trials):
    
    Q_arm=np.zeros(n_arms) # reward value for each arm
    N_visits = np.zeros(n_arms) #  arm  
    
    for j in range(n_steps):
        R_arm=np.random.normal(P_arm, 1, n_arms) # generate actual reward average probabilty +/- 1 
        
        if np.random.rand(1) < epsilon: 
            action = np.random.choice(n_arms, 1) #  exploration
        else:
            action = np.argmax(Q_arm) #   Reward   arm  Greedy action
            
            
        if N_visits[action] == 0:
            Q_arm[action] += R_arm[action] #  Q value
        else:
            Q_arm[action] += (R_arm[action] - Q_arm[action])/N_visits[action] #Q value update
            
            
        N_visits[action] += 1
        Result_array[j] += R_arm[action] #  trial Reward 
        
        Result_array2[j] += int(np.argmax(R_arm) == action) # count optimal action selection
    
plt.plot(Result_array2/n_trials, color='blue')      
```

```python
#Q3

n_arms = 10
n_steps = 2000
n_trials = 1000

epsilon_values = [0.1, 0.01] #epsilon  
c_values = [1, 1.5, 2, 4] #c  

#step = 0

class Bandit:
    def __init__(self, arm_index):
        self.arm_index = arm_index
        self.Q_arm = 0  # Initial reward value
        self.N_visits = 0  # arm  

    def update(self, reward):
        self.N_visits += 1
        self.Q_arm += (reward - self.Q_arm) / self.N_visits #Q update


def ucb_action_selection(bandits, c, step):
    #ucb . 0    (10^-6) 
    ucb_values = [bandit.Q_arm + c * np.sqrt(np.log(step + 1) / (bandit.N_visits + 10**(-6))) for bandit in bandits]
    action = np.argmax(ucb_values)
    return action


def epsilon_greedy_action_selection(bandits, epsilon):
    if np.random.rand(1) < epsilon:
        action = np.random.randint(n_arms) # epsilon selection
    else:
        action_values = [bandit.Q_arm for bandit in bandits]
        action = np.argmax(action_values) # greedy selection
    return action


def simulate_bandit_algorithm(action_selection_fn, param, epsilon=0):
    Result_array = np.zeros(n_steps)  # Result array for plotting optimal action selection
    P_arm = np.random.randn(n_arms)  # Average probability of reward amount
    
    for i in range(n_trials):
        bandits = [Bandit(j) for j in range(n_arms)] #Bandit 1~10 in list
        

        for step in range(n_steps):
            R_arm = np.random.normal(P_arm, 1, n_arms)  # Generate actual reward average probability +/- 1

            if action_selection_fn == ucb_action_selection: #function  
                action = ucb_action_selection(bandits, param, step)
            else:
                action = epsilon_greedy_action_selection(bandits, epsilon)

            bandit = bandits[action]
            reward = R_arm[action]
            bandit.update(reward)

            optimal_action = np.argmax(R_arm)
            Result_array[step] += int(optimal_action == action)

    return Result_array / n_trials



# UCB Action Selection
ucb_results = []
for c in c_values:
    ucb_result = simulate_bandit_algorithm(ucb_action_selection, c)
    ucb_results.append(ucb_result)

# Epsilon-Greedy Policy
epsilon_results = []
for epsilon in epsilon_values:
    epsilon_result = simulate_bandit_algorithm(epsilon_greedy_action_selection, epsilon)
    epsilon_results.append(epsilon_result)

# Plot
x_values = np.arange(n_steps)
plt.figure(figsize=(12, 6))

for i in range(len(c_values)):
    plt.plot(x_values, ucb_results[i], label=f'UCB (c={c_values[i]})')

for i in range(len(epsilon_values)):
    plt.plot(x_values, epsilon_results[i], label=f'Epsilon-Greedy (Epsilon={epsilon_values[i]})')

plt.xlabel('Steps')
plt.ylabel('Optimal Action Selection')
plt.legend()
plt.show()
```
